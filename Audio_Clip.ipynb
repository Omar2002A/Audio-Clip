```python
from pydub import AudioSegment
```

```python
#Download Audio
audio = AudioSegment.from_file("test.mp3")
```

```python
#Clip Length In Second
clip_length = 1000 #Second = 1000 MilliSeconds
```

```python
clips = []
for i in range(0,len(audio),clip_length):
  clip = audio[ i : i + clip_length]
  clips.append(clip)
```

```python
for i, clip in enumerate(clips):
  clip.export(f"output_{i}.mp3",format = "mp3")
```
